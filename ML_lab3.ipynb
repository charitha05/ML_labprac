```python
#Q1
import pandas as pd
from pandas import DataFrame
from pandas import Series
df=pd.read_csv('Automobile_data.csv',error_bad_lines=False)
print(df.head(5))
print(df.tail(5))
```

```python
#Q2
ma=max(df['price'])
x=df[df.price==ma]
print(x['company'])
```

```python
#Q3
print(df[df.company=='toyota'])
```

```python
#Q4
s=list(set(df['company']))
s.sort()
for i in s:
  print(i,len(df[df.company==i]))
```

```python
#Q5
for i in s:
  x=df[df.company==i]
  print(i,max(x['price']))
```

```python
#Q6
for i in s:
  x=df[df.company==i]
  print(i,x['average-mileage'].mean())
```
